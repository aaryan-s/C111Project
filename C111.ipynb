```
import statistics
import csv
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import random
#math = []
t = 0
integer = []
meanlist = []
dataframe = pd.read_csv("C111.csv")
math = dataframe["Math_score"].tolist() 
def rando():
    i = 0
    while i<100:
        x = math[random.randint(0,len(math)-1)]
        integer.append(x)
        i+=1
    average = statistics.mean(integer)
    meanlist.append(average)
while t<1000:
    rando()
    t+=1
fig = ff.create_distplot([meanlist],["Score"])
fig.show()
standard1 = statistics.stdev(math)
standard2 = statistics.mean(meanlist)
print("Math list stddev:",standard1)
print("Meanlist stddev:",standard2)

```
